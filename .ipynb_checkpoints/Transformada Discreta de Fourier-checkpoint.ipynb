### Aliasing

Supongamos que tenemos una onda _pura_, formada, digamos, por un seno de una Ãºnica frecuencia, por ejemplo: 
$$f_1(x) = \textrm{sen}(4\pi x).$$
Si recibimos una onda de este tipo y la digitalizamos, no vamos a obtener una onda sino un vector que evaluaciones de $f_1$ a ciertos intervalos de tiempo. La _frecuencia de sampleo_ es la frecuencia a la cual se capturan datos. Por ejemplo si tomamos 10 mediciones en un segundo, tenemos una _frecuencia de sampleo_ de 10 Hz (Hertz). 

Simulemos esto:


```python
def f1(x):
    return np.sin(4*np.pi*x)
```

```python
fs = 10
t = np.arange(fs)/fs
print(t)
# t tiene 10 casilleros que van de 0 a 1-1/10. 
```

```python
y1 = f1(t)
# y1 contine las mediciones de nuestra onda (10 mediciones en 1 seg.)
```

```python
plt.scatter(t,y1)
plt.plot(t,y1)
plt.title("k=2, con frecuencia de sampleo 10 Hz")
```

Consideremos ahora 
$$f_2(x) = \textrm{sen}(24\pi x),$$
medido con la misma frecuencia de sampleo.

```python
def f2(x):
    return np.sin(24*np.pi*x)
```

```python
y2 = f2(t)
plt.plot(t,y2)
plt.scatter(t,y2)
```

Luce igual, Â¿no? Â¿Por quÃ©?

Grafiquemos todo junto y agreguemos las ondas reales ($f1$ y $f2$) sobre una grilla mÃ¡s fina

```python
tt = np.linspace(0,1,1000)
plt.scatter(t,y1,label="mediciones k = 2")
plt.scatter(t,y2,label="mediciones k = 12")
plt.plot(tt,f1(tt),label="onda k = 2")
plt.plot(tt,f2(tt),label="onda k = 12")
plt.legend()
plt.title("Aliasing")
```

Lo que ocurre es que al tomar datos cada 1/10 de segundo, le pegamos justo a instantes en que las dos ondas son iguales. Es decir: para una frecuencia de sampleo de 10 Hz, la onda con frecuencia 2 es **indistinguible** de la onda con frecuencia 12. Esto, por supuesto, no es casual: 12 = 2 + 10. Lo mismo ocurrirÃ­a si tomamos la onda con frecuencia 22, 32, ... y tambiÃ©n con frecuencia -8, -18, etc. 

```python
K = 13
fig,ax = plt.subplots(K,1)
for k in range(0,K):
    ax[k].plot(t,np.sin(2*np.pi*k*t))
    
```

### Transformada discreta

```python
def TDF(x):
    N = len(x)
    l = np.arange(N)
    k = l.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * l / N)
    X = (1/N)*np.dot(e, x)
    return X
```

A modo de ejemplo, construyamos una seÃ±al y calculemos la transformada. Trabajemos con exponenciales complejas. Consideramos: 
$$f(x) = 2e^{2\pi i 8x} + 12 e^{2\pi i 12 x}$$

```python
def f(x):
    return 2*np.exp(2j*np.pi*8*x)+9*np.exp(2j*np.pi*12*x)
```

```python
plt.plot(tt,f(tt).real)
plt.title("Onda real")
```

Tomemos ahora un muestreo de nuestra onda, con frecuencia de sampleo 20 (para asegurarnos de que las dos frecuencias involucradas en $f$ sean distinguibles). 

```python
x = np.arange(20)/20
z = f(x)
zt = TDF(z) #transformada
z
```

```python
plt.plot(x,z.real)
```

Y grafiquemos la transformada.

```python
plt.scatter(np.arange(20),zt.real)
plt.title("Transformada discreta de z")
print(zt.real)
```

La transformada es $0$ para casi todas los valores salvo... en el casillero $8$, donde vale $2$ y en el casillero $12$, donde vale $9$. 
Es decir: la transformada nos estÃ¡ indicando quÃ© frecuencias estÃ¡n presentes en nuestra onda y con quÃ© amplitud. O, dicho mÃ¡s general: nos estÃ¡ dando los coeficientes de la escritura de nuestra onda _en las frecuencias que es posible distinguir con nuestra frecuencia de sampleo_. 

Otro ejemplo:

Consideremos ahora una onda real, por ejemplo: 
$$g(x) = 1 + 5\textrm{sen}(2\pi 3x) + 7\textrm{sen}(2\pi 5x).$$
Para anticipar lo que deberÃ­amos ver con la transformada, debemos pasar esto a exponenciales complejas: 

$$g(x) = 1 + \frac{5}{2i}(e^{i 2\pi 3 x}-e^{-i 2\pi 3 x}) + \frac{7}{2i} (e^{i 2\pi 5 x}-e^{-i 2\pi 5 x})$$

Construyamos ahora una onda real (con parte imaginaria nula):

```python
xx = np.arange(20)/20
g = lambda x: 1 + 5*np.sin(6*np.pi*x) + 7*np.cos(10*np.pi*x)
zz = g(xx)
xcont = np.linspace(0,1,1000)
zcont = g(xcont)
plt.plot(xcont,zcont)
plt.scatter(xx,zz)
```

```python
zzt = TDF(zz)
plt.scatter(np.arange(20),np.abs(zzt))
```

Se observa una simetrÃ­a en la transformada, en torno al centro del intervalo de frecuencias. Esto es consecuencia de la propiedad que dice que: 
$$\hat{x}_{N-k} = \hat{x}_k.$$
Pero las frecuencias _naturales_ no son las que estamos viendo. En lugar de la frecuencia 15 deberÃ­amos ver la frecuencia -5. SerÃ­a mejor poder hacer un grÃ¡fico con las frecuencias _correctas_.

### Usando el comando de Python

El comando `np.fft.fft` hace (casi) lo mismo que nuestra funciÃ³n `TDF` pero con un algoritmo muchÃ­simo mÃ¡s rÃ¡pido. 

**Importante:** Hay una diferencia entre `np.fft.fft` y `TDF`. `Numpy` no pone el factor $\frac{1}{N}$ en la definiciÃ³n de la transformada, sino en la definiciÃ³n de la inversa. Por lo tanto, al hacer `np.fft.fft` vamos a obtener coeficientes mÃ¡s grandes que con `TDF`. Para recuperar los coeficientes originales, hay que dividir por `N`.

```python
zzfft = np.fft.fft(zz)
N     = len(zzfft)
plt.scatter(np.arange(N),np.abs(zzfft))
#plt.scatter(np.arange(N),np.abs(zzfft)/N)
```

```python
zzfftshifteado = np.fft.fftshift(zzfft)
plt.scatter(np.arange(20),np.abs(zzfftshifteado))
```

Para ver las frecuencias originales, tenemos la funciÃ³n `fftfreq`. `fftfreq` recibe el nÃºmero de datos y el paso del sampleo y nos da las frecuencias que corresponden a un vector de tantos casilleros como querramos. Pero ademÃ¡s, nos da las frecuencias considerando las negativas. Por lo tanto, las podemos usar para graficar:

```python
frecuencias = np.fft.fftfreq(N,1/N)
```

```python
frecuencias
```

```python
plt.scatter(frecuencias,np.abs(zzfft))
```

### Jugando con audio

```python
import sounddevice as sd
```

```python
fs = 44100 # frecuencia de sampleo estandar
t  = np.arange(fs)/fs
y  = np.sin(2*np.pi*440*t) # frecuencia 440: LA en 4ta octava.
```

```python
sd.play(y,fs)
```

```python
y2 = np.sin(2*np.pi*440*t)\
     + 0.75*np.sin(2*np.pi*880*t)\
     + 0.25*np.sin(2*np.pi*440*3*t)\
     + 0.1*np.sin(2*np.pi*440*4*t)\
     + 0.02*np.sin(2*np.pi*440*5*t)\
     + 0.01*np.random.random(len(y))
```

```python
sd.play(y2,fs)
```

Pongamosle ruido a una nota:

```python
yr = y + 10*np.random.random(len(y))
```

```python
sd.play(yr,fs)
```

```python
yrt = np.fft.fft(yr)
freqs = np.fft.fftfreq(fs,1/fs)
```

```python
plt.scatter(freqs,np.abs(yrt))
```

```python
def matar_ruido(transf,umbral):
    for i in range(len(transf)):
        if np.abs(transf[i]) < umbral:
            transf[i] = 0.
    return transf
```

```python
yrtp = matar_ruido(yrt,5000)
```

```python
yrlimpia = np.fft.ifft(yrtp).real
```

```python
sd.play(yrlimpia,fs)
```
