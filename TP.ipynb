### Reconociendo notas (afinando)

El vector `notas` generado por el siguiente código contiene las notas musicales: `C`,`C#`,`D`,`D#`,`E`,`F`,`F#`,`G`,`G#`,`A`,`A#`,`B` (Do, Do sostenido, Re, Re sostenido, Mi, Fa, Fa sostenido, Sol, Sol sostenido, La, La sostenido, Si, respectivamente). Cada nota va acompañana de un número que indica en qué octava se encuentra la nota: La nota `A0` (La en primera octava) es igual, pero más grave que `A4` (La en cuarta octava)

```python
letras = ['C','C#','D','D#','E','F','F#','G','G#','A','A#','B']
notas_1 = []
for i in range(9):
    for l in letras:
        notas_1.append(l+str(i))
notas_1 = np.array(notas_1)
```

El siguiente vector contiene las frecuencias que corresponden a cada una de estas notas:

```python
frecuencias_notas = 16.3516*(2**(1/12))**np.arange(len(notas_1))
```

De este modo, si queremos encontrar qué frecuencia le corresponde a la nota `A4` podemos hacer:

```python
indice_A4 = np.nonzero(notas_1=='A4')[0][0] #esto da el índice de notas donde está A4.
print("Frecuencia de A4: ",frecuencias_notas[indice_A4], "Hz")
```

#### Ejercicio: 
Los archivos `notas*.wav` contienen grabaciones de distintas notas musicales tocadas con diferentes instrumentos. Se quiere identificar qué nota contiene cada archivo. 
Para ello, leer el archivo y convertirlo en un array de numpy, calcularle la transformada de Fourier, graficarla y estudiar el gráfico.

```python
fs,nota1 = wav.read("Notas/Notas/nota1.wav")
fs,nota2 = wav.read("Notas/Notas/nota2.wav")
fs,nota3 = wav.read("Notas/Notas/nota3.wav")
fs,nota4 = wav.read("Notas/Notas/nota4.wav")
fs,nota5 = wav.read("Notas/Notas/nota5.wav")
fs,nota6 = wav.read("Notas/Notas/nota6.wav")
sd.play(nota2,fs)
```

```python
notas=[nota1,nota2,nota3,nota4,nota5,nota6] ##no nos dejaba hacer un array con los datos porque tenian distintas dimensiones
```

```python
notas = np.array(notas, dtype=object) ##transformamos la lista en un array
##calculo la transformada de fourier
r=[]
n=[]
for nota in notas:
    N=len(nota)
    y=(1/N)*fft(nota)[:N // 2] ##[:N // 2] toma la mitad de los datos ya que la transformada es simétrica, asi que tomo la mitad positiva
    r.append(y)
    n.append(N)
```

```python
##NOTA 1
sd.play(nota1,fs)
plt.plot(nota1)
plt.show() ##grafico de la notal archivo
y1=r[0]
plt.plot(y1[:1000],label="nota 1")
plt.xlim(600,1000)
plt.legend() ##grafico de la transformada del archivo
plt.show()
frecuencia1 = np.fft.fftfreq(n[0], 1 / fs)[:n[0] // 2] ## genero un array de frecuencias correspondientes a los valores de la transformada de fourier

indice_frecuencia_principal1 = np.argmax(y1) ##encuentro el indice con mayor magnitud de y1
frecuencia_principal1 = frecuencia1[indice_frecuencia_principal1] ##encuentro la frecuencia que corresponde al índice de mayor magnitud
print("Frecuencia principal:", frecuencia_principal1, "Hz")

def frecuencia_a_nota(frecuencia):
    A4 = 440  # Frecuencia de A4 en Hz
    
    # Calcular la cantidad de semitonos desde A4 y redondear al entero más cercano
    semitonos_desde_A4 = int(round(12 * np.log2(frecuencia / A4)))
    
    # Lista de notas en una octava
    notas = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]
    
    # Calcular la nota y la octava a partir de los semitonos desde A4
    nombre_nota = notas[(semitonos_desde_A4 + 9) % 12]  # +9 para ajustar a C como 0 en la lista
    octava = 4 + (semitonos_desde_A4 + 9) // 12  # 4 es la octava de A4

    return f"{nombre_nota}{octava}"
print("La nota 1 corresponde a",frecuencia_a_nota(frecuencia_principal1))
```

### Limpiando audio
Los archivos `secret*.wav` contienen fragmentos de audio tapados con ruido. Queremos limpiar el ruido de modo de lograr oír el mensaje oculto. El ruido es diferente en cada archivo y por lo tanto el proceso de limpieza debe ser artesanal, haciendo algo ligeramente distinto en cada caso. 

Para cada archivo:

1. Leer cada archivo para obtener un vector de Numpy. Ejecutarlo para escuchar la grabación. 

**Cuidado:** En algunos audios, el ruido es un pitido **muy** agudo. No usar auriculares o usarlos fuera del oído. 

2. Calcular la transformada de Fourier del vector y graficar el resultado. 

3. A partir del gráfico, elaborar una estrategia y limpiar las frecuencias que se supone que producen el ruido. Luego, aplicar la transformada inversa al vector de transformadas procesado y ejecutar el audio resultante. 

**Nota:** Es posible que luego de limpiar el ruido el audio quede a muy bajo volumen. Para subir el volumen, basta con multiplicar el vector resultante por un factor mayor que 1. Por ejemplo: al multiplicar por 2 se obtiene el mismo sonido con el doble de volumen. 

```python
fs,secreto1 = wav.read("secrets/secret1.wav")
fs,secreto2 = wav.read("secrets/secret2.wav")
fs,secreto3 = wav.read("secrets/secret3.wav")
fs,secreto4 = wav.read("secrets/secret4.wav")
fs,secreto5 = wav.read("secrets/secret5.wav")
secretos=[secreto1,secreto2,secreto3,secreto4,secreto5]
secretos= np.array(secretos, dtype=object)
for secret in secretos:
    sd.play(secret,fs)
    time.sleep(5)
```

```python
secreto_fft1 = fft(secreto1)
frecuencias1 = np.fft.fftfreq(len(secreto_fft1), 1/fs)
frecuencias_positivas1 = frecuencias1[:len(frecuencias1)//2]
secreto_fft_positivo1 = np.abs(secreto_fft1)[:len(frecuencias1)//2]
plt.figure(figsize=(10, 6))
plt.plot(frecuencias_positivas1,secreto_fft_positivo1 )
plt.xlim(0,1500)
plt.title("Espectro de frecuencias de la señal original")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid()
plt.show()

umbral1 = 1000  # Frecuencia a partir de la cual consideramos que el ruido está presente, se puede observar en el gráfico al aplicar xlim

# Suprimir frecuencias fuera del rango de la nota musical
# Vamos a atenuar las frecuencias de mayor a 1000 Hz (aproximadamente) 
secreto1_fft_limpio = secreto_fft1.copy()
secreto1_fft_limpio[np.abs(frecuencias1) > umbral1] = 0  ##elimino frecuencias mayores al umbral
secreto1_limpio_positivo = np.abs(secreto1_fft_limpio)[:len(frecuencias1)//2] ##tomo los valores positivos
plt.figure(figsize=(10, 6))
plt.plot(frecuencias_positivas1 ,secreto1_limpio_positivo  )
plt.title("Espectro de frecuencias después de la supresión de ruido")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

secreto_1_sinruido = np.fft.ifft(secreto1_fft_limpio) ##convierto una señal que está en el dominio de la frecuencia de regreso al dominio del tiempo.
t = np.linspace(0, N / fs, N) ##genero un linspace de tiempo para graficar
# Graficar la señal reconstruida
plt.figure(figsize=(10, 6))
plt.plot(t[:1000], secreto_1_sinruido[:1000])  # Graficar una parte de la señal
plt.title("Señal limpia")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
secreto_1_final = np.real(secreto_1_sinruido).astype(np.float32)  # Convertir a float32 y tomar solo la parte real
##necesito tomar los valores reales ya que la antitransformada devuelve algunos valores imaginarios
##astype convierte mis datos a tipo de datos float32, que se utiliza para audio

# Factor de amplificación 
factor_amplificacion = 5.0 ##aumenta el volumen

# Amplificar el audio
secreto1_amplificado = secreto_1_final * factor_amplificacion

# Normalizar si es necesario para mantener el audio en el rango [-1, 1]
secreto1_amplificado /= np.max(np.abs(secreto1_amplificado))

# Reproducir el audio amplificado
sd.play(secreto1_amplificado, samplerate=fs)
sd.wait() 
##es you talking to me
```
